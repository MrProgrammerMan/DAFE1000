# Numerisk integrasjon av polynom
## Teori
### Polynom
Et polynom er på formen a0 + a1 * x + a3 * x^2 ... + an * x^n, hvor n er et naturlig tall som kalles polynomets grad.

### Derivasjon & integrasjon
Derivasjon vil si å bestemme endringen i funksjonsverdien over variablen, i dette tilfellet x. Det motsatte av derivasjon kalles for antiderivasjon eller integrasjon.

For en funksjon s(t) som viser strekning tilbakelagt av en bil ved tid t, er den deriverte s'(t) endringen i avstand, eller altså farten til bilen ved tid t. Videre er s''(t) aksellerasjonen til bilen ved tid t.

## Teoretisk løsning
For mange funksjoner er den deriverte og integrerte bestemt og enkel å regne ut. For eksempel er den deriverte av (x^2)' = 2x og på sammen måte integral(2x) = x^2.

Dette er likevel ikke tilfellet for alle funksjoner, og noen ganger er det svært vanskelig eller umulig å finne en nøyaktig løsning.

## Numerisk løsning
En numerisk løsning handler om å bruke logiske metoder for å løse et problem ved hjelp av programmering. Vi kan for eksempel velge å dele opp en funksjon i bestemte intervaller og på den måten behandle en forenkler versjon av funksjonen for å finne en *tilnærmet* løsning. Vi **diskreteserer** funksjonen.

Ved å benytte moderne programmering kan vi dele en graf inn i mange veldig små deler, som lar oss estimere integralet av funksjonen til en svært nøyaktig grad.

Koden under plotter en graf f(x) fra 0 til 10.

```python
import matplotlib.pyplot as plt

lower_bound = 0
upper_bound = 10
step = 0.1

def f(x):
  return 4*x**3 - 3*x**2 + 2*x - 1

x_values = []
i = lower_bound
while (i < upper_bound):
  x_values.append(i)
  i += step

y_values = []
for value in x_values:
  y_values.append(f(value))

plt.plot(x_values, y_values)

```

Hvis vi nå ønsker å beregne integralet av funksjonen numerisk, kan vi dele opp funksjonen i diskrete deler.

```python
parts = 5
part_width = (upper_bound - lower_bound) / parts

integral_sum = 0

for i in range(parts):
  x = lower_bound + i * part_width
  integral_sum += f(x) * part_width

# Visually show the integral
for i in range(parts):
  x = lower_bound + i * part_width
  y = f(x)
  plt.fill_between([x, x + part_width], [y, y], color='gray', alpha=0.5)

plt.plot(x_values, y_values)
plt.title(f"Integral: {integral_sum}")
plt.show()
```

Eksempelet over benytter det som kalles for nedre sum. Her deler vi funksjoner opp i rektangler under grafen.
Vi kan øke antallet rektangler for høyere nøyaktighet:

```python
parts = 100
part_width = (upper_bound - lower_bound) / parts

integral_sum = 0
for i in range(parts):
  x = lower_bound + i * part_width
  integral_sum += f(x) * part_width

# Visually show the integral
for i in range(parts):
  x = lower_bound + i * part_width
  y = f(x)
  plt.fill_between([x, x + part_width], [y, y], color='gray', alpha=0.5)

plt.plot(x_values, y_values)
plt.title(f"Integral: {integral_sum}")
plt.show()
```

Vi har også øvre sum:

```python
parts = 5
part_width = (upper_bound - lower_bound) / parts

integral_sum = 0
for i in range(parts):
  x = lower_bound + i * part_width + part_width
  integral_sum += f(x) * part_width

# Visually show the integral
for i in range(parts):
  x = lower_bound + i * part_width
  y = f(x + part_width) + part_width
  plt.fill_between([x, x + part_width], [y, y], color='gray', alpha=0.5)

plt.plot(x_values, y_values)
plt.title(f"Integral: {integral_sum}")
plt.show()
```

Og mer nøyaktig kan vi beregne gjennomsnittlig sentrumssum:

```python
parts = 5
part_width = (upper_bound - lower_bound) / parts

integral_sum = 0
for i in range(parts):
  x = lower_bound + i * part_width + part_width / 2
  integral_sum += f(x) * part_width

# Visually show the integral
for i in range(parts):
  x = lower_bound + i * part_width
  y = f(x + part_width / 2) + part_width
  plt.fill_between([x, x + part_width], [y, y], color='gray', alpha=0.5)

plt.plot(x_values, y_values)
plt.title(f"Integral: {integral_sum}")
plt.show()
```

Ved å bruke smmå intervaller kan dette gi svært nøyaktig resultater, tilpasset den nødvendige nøyaktigheten i forskjellige situasjoner:

```python
parts = 500
part_width = (upper_bound - lower_bound) / parts

integral_sum = 0
for i in range(parts):
  x = lower_bound + i * part_width + part_width / 2
  integral_sum += f(x) * part_width

# Visually show the integral
for i in range(parts):
  x = lower_bound + i * part_width
  y = f(x + part_width / 2) + part_width
  plt.fill_between([x, x + part_width], [y, y], color='gray', alpha=0.5)

plt.plot(x_values, y_values)
plt.title(f"Integral: {integral_sum}")
plt.show()
```
