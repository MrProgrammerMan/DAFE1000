```python
import numpy as np
import matplotlib.pyplot as plt

x = float(input("Real part: "))
y = float(input("Imaginary part: "))

n = int(input("Natural number: "))

r = np.power(x**2+y**2, 1/(2*n))

theta = np.atan(y/x)

sols_x = []
sols_y = []

for k in range(0, n):
  ang = (theta+2*np.pi*k)/n
  sols_x.append(r*np.cos(ang))
  sols_y.append(r*np.sin(ang))

plt.scatter(sols_x, sols_y, color='blue')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```
